```python
#importing libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
#using pandas to read the database stored in the same folder
data=pd.read_csv("mnist_train.csv")
data1=pd.read_csv("mnist_test.csv")
```

```python
#viewing colum heads
data.head()
```

```python
#reshaping extracted data
a=a.reshape(28,28).astype('uint8')
plt.imshow(a)
plt.show()
```

```python
#preparing the data
#separating labels and data values
df_x=data.iloc[:,1:]
df_y=data.iloc[:,0]
```

```python
#creating train and test sizes
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=4)
```

```python
#checking data 
x_train.head()
```

```python
#checking data 
y_train.head()
```

```python
#call rf_classifier
rf=RandomForestClassifier(n_estimators=100)
```

```python
#fitting model
rf.fit(x_train, y_train)
```

```python
#prediction on test data
pred=rf.predict(x_test)
print(pred)
```

```python
#checking precition accuracy
s=y_test.values
#calculating number of correctly predicted values
count=0
for i in range(len(pred)):
    if pred[i]==s[i]:
        count+=1

print(count)
```

```python
#total number of values
print(len(pred))
#accuracy
acc=count/len(pred)
print("Model Accuracy using Random Forest Classifier is=",round(acc*100,4),end="%")
```
